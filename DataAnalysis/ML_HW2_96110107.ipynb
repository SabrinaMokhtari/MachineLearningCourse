```python
def find_max_item(series):
    return series.count()
df1 = df[df['Risk'] == 'good'][['Purpose', 'Sex', 'Risk']].groupby(['Purpose', 'Sex']).agg(find_max_item)
df2 = df[df['Risk'] == 'bad'][['Purpose', 'Sex', 'Risk']].groupby(['Purpose', 'Sex']).agg(find_max_item)
con_df = pd.concat([df1, df2], axis=1, join='inner')
con_df.columns = ['Good Risk', 'Bad Risk']
con_df = con_df.reset_index()
con_df = con_df[con_df['Good Risk'] > 2 * con_df['Bad Risk']]
con_df['Good Bad Difference'] = con_df['Good Risk'] - con_df['Bad Risk']
for c in con_df['Purpose'].unique():
    plt.figure(figsize=(5, 8))
    sns.countplot(x="Good Bad Difference", hue="Sex", data=con_df[con_df['Purpose'] == c]);
```

<div style="direction:rtl"> برای مثال ، هنگامی که ما تحلیل انجام می دهیم، مانند multivariate linear regression، متغیر که داده های در رنج بالا تری دارد ، نتیجه را بیشتر تحت تأثیر قرار می دهد. اما این لزوماً به این معنی نیست که به عنوان predictor اهمیت بیشتری دارد. بنابراین داده ها را normalize می کنیم تا همه متغیرها در یک محدوده قرار بگیرند. همچنین این مسئله باعث بالا رفتن زمان اجرا و train می شود و مدت زمان همگرایی ضرایت به مقدار درست را بیشتر می کند</div>

```python
val = new_df.values
min_max_scaler = preprocessing.MinMaxScaler()
val_scaled = min_max_scaler.fit_transform(val)
new_df = pd.DataFrame(val_scaled)
new_df
```

<div dir=rtl>
    ۲۰.(۴ نمره)
مقدار و بردار ویژه‌ی ماتریس کواریانس را بدست بیاورید.
</div>

```python
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)
eigen_values, eigen_vectors
```

<div dir=rtl>
    ۲۱.(۳ نمره)
    برای انتخاب اجزای اساسی(Principle Components) سه کار ریز را انجام دهید:
    <br\>
هر بردار ویژه را با مقدار ویژه متناظرش مربوط کنید (می‌توانید بصورت جفتیpair آن‌ها را کنار هم قرار دهید)
        <br\>
            لیستی که بردار و مقدار ویژه‌ها دارید را بر اساس مقدار ویژه‌ها مرتب کنید
            <br\>
                مقدار ویژه‌های مرتب شده را به ترتیب نزولی در خروجی نمایش دهید
</div>

```python
value_vector_pair = []
for x, y in zip(eigen_values, eigen_vectors):
    value_vector_pair.append((x, y))
value_vector_pair
```

```python
sorted_value_vector_pair = sorted(value_vector_pair, key=lambda x: x[0], reverse=True)
sorted_value_vector_pair
```

```python
sorted_eigenvalues = [sorted_value_vector_pair[i][0] for i in range(len(sorted_value_vector_pair))]
sorted_eigenvectors = [list(sorted_value_vector_pair[i][1]) for i in range(len(sorted_value_vector_pair))]
sorted_eigenvalues, sorted_eigenvectors
```

<div dir=rtl>
    ۲۲.(۴ نمره)
همه‌ی
    explained variance 
    را بدست بیاورید و در نمودار pie نمایش دهید
    حتما درصد هرکدام از اسلایس‌ها در نمودار مشخص شده‌باشد.
</div>

```python
total_eigen_values = sum(eigen_values)
var_exp = [(i/total_eigen_values) for i in sorted(eigen_values, reverse=True)]
fig1, ax1 = plt.subplots()
ax1.pie(var_exp, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
